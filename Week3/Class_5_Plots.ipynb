### Data distribution. displot

```python
sns.displot(data=df, x="Wing", col="Taxon", kde=True)
```

```python
sns.displot(data=df, x="Wing", col="Taxon", hue="Island", rug=True)
```

Or the other way round...


```python
sns.displot(data=df, x="Wing", col="Island", hue="Taxon", rug=True)
```

### Categorical data. catplot

```python
sns.catplot(data=df, kind="swarm", x="Taxon", y="Wing", hue="Island")
```

```python
### realtionships. relplot
```

```python
sns.relplot(
    data=df,
    x="Wing", y="Tail", col="Taxon")
```

```python
linear regression plots
```

```python
sns.lmplot(data=df, x="Wing", y="Tail", col="Taxon")
```

### Composite plots

```python
sns.jointplot(data=df, x="Wing", y="Tail", hue="Taxon")
```

```python
sns.jointplot(data=df, x="Wing", y="Tail", hue="New_Taxonomy")
```
